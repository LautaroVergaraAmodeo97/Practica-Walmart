Ahora vamos a realizar un promedio de cierre (Close) por cada aÃ±o y vamos a graficar en un plot lineal.

```python
promedio_cierre = df.groupby('Year')['Close'].mean().reset_index()

promedio_cierre
```

```python
#Configuramos el plot
fig,ax = plt.subplots(figsize=(6,4))

sns.lineplot(x=promedio_cierre['Year'],y=promedio_cierre['Close'],data=promedio_cierre)

plt.xlabel('AÃ±o')
plt.ylabel('Promedio de cierre')
plt.title('Cierre de acciones por cada aÃ±o')
plt.show()
```

Ahora vamos a hacer un grÃ¡fico de barras pero de los ultimos 5 aÃ±os (2020-2025)

```python
ultimos_5_aÃ±os = promedio_cierre.tail(5)

ultimos_5_aÃ±os
```

```python
#Configuramos el plot
fig,ax = plt.subplots(figsize=(6,4))

sns.barplot(data=ultimos_5_aÃ±os,x=ultimos_5_aÃ±os['Year'],y=ultimos_5_aÃ±os['Close'])

plt.xlabel('AÃ±o')
plt.ylabel('Promedio de cierre')
plt.title('Cierre de acciones por cada aÃ±o')
plt.show()
```

La primera observaciÃ³n que podemos notar es que 2021 y 2022 estÃ¡n practicamente igualado. Esto es debido a post pandemia de COVID 19 y que el gobierno de turno estadounidense no creciÃ³ en esos aÃ±os.

Ahora vamos a realizar un histograma de los precios de cierre de los Ãºltimos 5 aÃ±os.

```python
#Configuramos el plot
fig,ax = plt.subplots(figsize=(15,6))

sns.histplot(data=promedio_cierre,x=promedio_cierre['Year'],y=promedio_cierre['Close'],bins=5)

plt.xlabel('AÃ±o')
plt.ylabel('Promedio de cierre')
plt.title('Cierre de acciones por cada aÃ±o')
plt.show()
```
